```python
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg?crop=0.752xw:1.00xh;0.175xw,0&resize=1200:*')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
import cv2
```

```python
img = cv2.imread('/content/dog.jpg?crop=0.752xw:1.00xh;0.175xw,0&resize=1200:*')
```

```python
type(img)
```
